## Load Data to Mysql Tables

```python
def populate_table(db_connection, db_cursor, insert_query, file_path):
    
    with open(file_path, mode='r') as csv_data:
        reader = csv.reader(csv_data, delimiter=';')
        csv_data_list = list(reader)
        for row in csv_data_list[1:]:
            row = tuple(map(lambda x: None if x == "" else x, row[0].split(',')))
            db_cursor.execute(insert_query, row)
        
    db_connection.commit()
```

## Question 3 (5 pts) 

Implement the following function: `get_payment`. It takes as input two parameters: startyear and endyear. It needs to find: Between the years startyear and endyear (startyear < year < endyear), how many times was each payment method used? Use `order_delivered_customer_date` of the ORDERS table to find the year of the order. 

Example result: `(credit_card, 215 times), (voucher, 20 times), (debit card, 2 times),` ...

HINT: There exists a MySQL function which makes it easy to extract years from dates.

```python
def get_payment(startyear, endyear):
    date = f"""
        SELECT op.payment_type, COUNT(*) AS payment_count
        FROM ORDER_PAYMENTS AS op, ORDERS AS o
        WHERE op.order_id = o.order_id AND YEAR(o.order_delivered_customer_date) BETWEEN {int(startyear) + 1} AND {int(endyear) - 1}
        GROUP BY op.payment_type
        ORDER BY payment_count DESC;
        """
    db_cursor.execute(date)
    results = db_cursor.fetchall()
    formatted_results = [(payment_type, f'{count} times') for payment_type, count in results]
    return formatted_results
```

```python
print(get_payment("2015", "2017"))
```

```python
import sqlalchemy as db
import pymysql
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy import inspect
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from sqlalchemy import and_
from datetime import datetime
from sqlalchemy import or_
from sqlalchemy.orm import Query
from sqlalchemy.sql.expression import func
from sqlalchemy.sql.expression import alias
```

```python
engine = db.create_engine("mysql+pymysql://root:@localhost/ps_ecommerce")
# engine = db.create_engine("mysql+pymysql://root:WRITE_YOUR_PASSWORD@localhost/ps_ecommerce")
connection = engine.connect()
```

```python
Base = declarative_base()
Base.metadata.reflect(engine, views = True)
```

```python
class Customers(Base):
    __table__ = Base.metadata.tables["CUSTOMERS"]
    __mapper_args__ = {
        'primary_key':[Base.metadata.tables["CUSTOMERS"].c.customer_id]
    }
    
class Orders(Base):
    __table__ = Base.metadata.tables["ORDERS"]
    __mapper_args__ = {
        'primary_key':[Base.metadata.tables["ORDERS"].c.order_id]
    }
    
class OrderItems(Base):
    __table__ = Base.metadata.tables["ORDER_ITEMS"]
    __mapper_args__ = {
        'primary_key':[Base.metadata.tables["ORDER_ITEMS"].c.order_id, Base.metadata.tables["ORDER_ITEMS"].c.order_item_id]
    }
    
class OrderPayments(Base):
    __table__ = Base.metadata.tables["ORDER_PAYMENTS"]
    __mapper_args__ = {
        'primary_key':[Base.metadata.tables["ORDER_PAYMENTS"].c.order_id, Base.metadata.tables["ORDER_PAYMENTS"].c.payment_sequential]
    }
    
class Products(Base):
    __table__ = Base.metadata.tables["PRODUCTS"]
    __mapper_args__ = {
        'primary_key':[Base.metadata.tables["PRODUCTS"].c.product_id]
    }
```

```python
class OrderCosts(Base):
    __table__ = Base.metadata.tables["ORDER_COSTS"]
    __mapper_args__ = {
        'primary_key':[Base.metadata.tables["ORDER_COSTS"].c.order_id]
    }
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

Consider a seller whose main product is 'cama_mesa_banho'. In order to increase his sales, he wants to make a special discount. If you buy 'cama_mesa_banho', you will get the item 'x' at half of its original price.

What should be this item 'x'? Explain the logic behind your reasoning, and find the item using SQL Alchemy ONLY.

The ideal item 'x' for a discount should be a product that customers often buy along with 'cama_mesa_banho'. The reason behind this choice is straightforward: people who are buying bed and bath linens might also be interested in related items, like pillowcases or towels. By offering a discount on an item that's frequently bought with 'cama_mesa_banho', the seller can encourage customers to buy more, increasing overall sales. To find this item 'x', we look at past sales data to see which products are usually bought with 'cama_mesa_banho' and choose the most common one for the discount. This method ensures that the choice is based on what customers actually tend to buy together, making it a smart and data-driven decision.

```python
# Write your solution here
#
#
#
#
subquery = session.query(OrderItems.__table__.c.order_id).join(
    Products, Products.__table__.c.product_id == OrderItems.__table__.c.product_id
).filter(Products.__table__.c.product_category_name == 'cama_mesa_banho').subquery()

query = session.query(
    Products.__table__.c.product_category_name,
    func.count(Products.__table__.c.product_id).label('frequency')
).join(
    OrderItems, Products.__table__.c.product_id == OrderItems.__table__.c.product_id
).filter(
    OrderItems.__table__.c.order_id.in_(subquery),
    Products.__table__.c.product_category_name != 'cama_mesa_banho'
).group_by(Products.__table__.c.product_category_name).order_by(func.count(Products.__table__.c.product_id).desc())

top_product = query.first()

print(f"The ideal item 'x' for the discount is: {top_product[0]} (Frequency: {top_product[1]})")
```
